```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
churn_data = pd.read_csv('../../output/socialnet_dataset.csv', index_col=[0, 1])

# plot the correlation scatter plots
sns.pairplot(churn_data, kind='scatter')
```

```python
# calculate the correlation matrix for the churn dataset

# sort columns alphabetically
churn_data = churn_data.reindex(sorted(churn_data.columns), axis=1)
# calculate Pearson correlation coefficient between metrics
corr_df = churn_data.corr()
```

```python
# create a heatmap from the correlation coefficients
sns.heatmap(corr_df, 
            annot=True, 
            fmt='.2f', 
            cmap='crest', 
            cbar=False,
            square=False)
```

```python
# save the correlation matrix for later
corr_df.to_csv('../../output/socialnet_dataset_correlation_matrix.csv')
```
