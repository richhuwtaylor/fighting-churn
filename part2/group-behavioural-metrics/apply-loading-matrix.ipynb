```python
import pandas as pd
import numpy as np
import os
```

Apples the loading matrix `socialnet_dataset_scores_load_mat.csv` to the dataset of scores to create the grouped average scores for each account-date observation.

```python
# Load the metric scores
score_data = pd.read_csv('../../output/socialnet_dataset_scores.csv', index_col=[0,1])

# Remove churn indicator for now
data_to_group = score_data.drop('is_churn', axis=1)

# Read in the loading matrix
load_mat_df = pd.read_csv('../../output/socialnet_dataset_scores_load_mat.csv', index_col=0)
load_max_ndarray = load_mat_df.to_numpy()

# Rearrange columns to the same order of the loading matrix rows
ndarray_to_group = data_to_group[load_mat_df.index.values].to_numpy() 
grouped_ndarray = np.matmul(ndarray_to_group, load_max_ndarray)

# Create a DataFrame from the npdarray result
churn_data_grouped = pd.DataFrame(grouped_ndarray,
                                  columns = load_mat_df.columns.values,
                                  index=score_data.index)

# Add back the churn indicator
churn_data_grouped['is_churn'] = score_data['is_churn']
```

```python
# Save the result
churn_data_grouped.to_csv('../../output/socialnet_dataset_groupscore.csv', header=True)
```
