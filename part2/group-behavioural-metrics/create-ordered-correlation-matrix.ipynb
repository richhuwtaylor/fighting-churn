```python
import pandas as pd
import os
```

This notebook crates a version of the correlation matrix whose columns appear in the same order as the rows of loading matrix created for grouping metrics together.

```python
# Load the metric scores
churn_data = pd.read_csv('../../output/socialnet_dataset_scores.csv', index_col=[0,1])

# Load the loading matrix
load_mat_df = pd.read_csv('../../output/socialnet_dataset_scores_load_mat.csv', index_col=0)

# Reorder the dataset columns to the order of the loading matrix rows
churn_data = churn_data[load_mat_df.index.values]

# Calculate the matrix of Pearson correlation coefficients
corr = churn_data.corr()
```

```python
# Save the result to disk
corr.to_csv['../../output/socialnet_dataset_ordered_correlation_matrix.csv']
```
